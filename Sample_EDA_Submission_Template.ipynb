## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# 1Ô∏è Drop duplicates
print("Initial dataset shape:", data.shape)
data.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", data.shape)

# 2Ô∏è Handle missing values
# Drop rows missing essential columns
data.dropna(subset=['title', 'type', 'release_year'], inplace=True)

# Fill missing categorical values
data['country'] = data['country'].fillna('Unknown')
data['rating'] = data['rating'].fillna('Not Rated')
data['director'] = data['director'].fillna('Unknown')
data['cast'] = data['cast'].fillna('Unknown')

# 3Ô∏è Convert data types
# Convert release_year to integer if not already
data['release_year'] = data['release_year'].astype(int)

# Convert date_added to datetime
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')

# 4Ô∏è Extract numeric features from duration
# For Movies: "90 min" ‚Üí 90
data['duration_minutes'] = data['duration'].str.replace(' min', '', regex=False)
data['duration_minutes'] = pd.to_numeric(data['duration_minutes'], errors='coerce')

# For TV Shows: "3 Seasons" ‚Üí 3
data['num_seasons'] = data['duration'].str.replace(' Season', '', regex=False)
data['num_seasons'] = data['num_seasons'].str.replace('s', '', regex=False)
data['num_seasons'] = pd.to_numeric(data['num_seasons'], errors='coerce')

# 5Ô∏è Derived feature: description length
data['description_length'] = data['description'].astype(str).apply(len)

# 6Ô∏è Clean categorical columns (strip whitespace)
for col in ['type', 'rating', 'country', 'listed_in']:
    data[col] = data[col].astype(str).str.strip()

# 7Ô∏è Check final missing values
print("\nMissing values after preprocessing:")
print(data.isnull().sum())

# 8Ô∏è Fill remaining missing numeric features with median
data['duration_minutes'].fillna(data['duration_minutes'].median(), inplace=True)
data['num_seasons'].fillna(data['num_seasons'].median(), inplace=True)



# 9Ô∏è Dataset ready for analysis
print("\n‚úÖ Dataset is now clean and analysis-ready!")
print("Final dataset shape:", data.shape)
display(data.head())
````

### What all manipulations have you done and insights you found?

After performing data wrangling on the Netflix dataset, several key manipulations were done to make it analysis-ready. Duplicate rows were removed to ensure data integrity, and missing values in essential columns such as title, type, and release_year were dropped, while missing categorical values in columns like country, director, cast, and rating were filled with placeholder values. Text-based numeric information, such as movie durations (duration) and TV show seasons, was converted into numeric columns (duration_minutes and num_seasons), and a derived feature description_length was created to quantify the length of each content description. Additionally, categorical columns were cleaned by stripping whitespace, and remaining numeric missing values were filled with median values. These manipulations provide several insights: the dataset now clearly distinguishes between movies and TV shows, allows quantitative analysis of content length and description complexity, reveals content distribution across countries and genres, and is fully structured for exploratory data analysis, univariate, bivariate, and multivariate studies, as well as clustering tasks.

#### Chart - 6

````python
# Chart - 6 visualization code
# Histogram with KDE
try:
    # Check if 'duration' and 'type' columns exist
    if 'duration' not in data.columns or 'type' not in data.columns:
        raise KeyError("The dataset must contain 'duration' and 'type' columns.")

    # Filter only movies (exclude TV Shows)
    movies = data[data['type'] == 'Movie'].copy()

    # Extract numeric duration (e.g., '90 min' -> 90)
    movies['duration_minutes'] = (
        movies['duration']
        .str.replace(' min', '', regex=False)
        .str.extract('(\d+)')
        .astype(float)
    )

    # Drop missing or invalid duration values
    movies = movies.dropna(subset=['duration_minutes'])

    # Plot histogram with KDE (smooth density curve)
    plt.figure(figsize=(10,6))
    plt.hist(
        movies['duration_minutes'],
        bins=30,
        density=True,
        color='lightcoral',
        edgecolor='black',
        alpha=0.7,
        label='Histogram'
    )

    # Kernel Density Estimate (KDE) plot
    from scipy.stats import gaussian_kde
    kde = gaussian_kde(movies['duration_minutes'])
    x_vals = np.linspace(movies['duration_minutes'].min(), movies['duration_minutes'].max(), 200)
    plt.plot(x_vals, kde(x_vals), color='darkred', linewidth=2, label='KDE Curve')

    plt.title("Distribution of Movie Durations on Netflix")
    plt.xlabel("Duration (minutes)")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(alpha=0.4, linestyle='--')
    plt.tight_layout()
    plt.show()

    # Display summary statistics
    print("üé¨ Summary Statistics for Movie Durations:")
    print(movies['duration_minutes'].describe())

except KeyError as e:
    print("‚ùå Error:", e)
except Exception as e:
    print("‚ùå An unexpected error occurred:", e)
````

##### 1. Why did you pick the specific chart?

````python
# Chart - 9 visualization code
# Grouped Bar chart
try:
    # Check for necessary columns
    required_cols = ['type', 'rating', 'duration']
    for col in required_cols:
        if col not in data.columns:
            raise KeyError(f"Required column '{col}' is missing.")

    # Separate Movies and TV Shows
    movies = data[data['type'] == 'Movie'].copy()
    tv_shows = data[data['type'] == 'TV Show'].copy()

    # For TV Shows, extract number of seasons
    tv_shows['num_seasons'] = (
        tv_shows['duration']
        .str.replace(' Season', '', regex=False)
        .str.replace('s', '', regex=False)
        .astype(float)
    )

    # Group Movies by Rating
    movie_counts = movies['rating'].value_counts().sort_index()

    # Group TV Shows by Rating (sum across all seasons)
    tv_counts = tv_shows.groupby('rating')['num_seasons'].sum()

    # Combine into a DataFrame
    grouped_data = pd.DataFrame({
        'Movies': movie_counts,
        'TV Shows': tv_counts
    }).fillna(0)

    # Plot grouped bar chart
    grouped_data.plot(kind='bar', figsize=(12,6), color=['skyblue', 'lightgreen'])
    plt.title("Rating Distribution of Movies vs TV Shows (TV Shows weighted by number of seasons)", fontsize=14, weight='bold')
    plt.xlabel("Content Rating")
    plt.ylabel("Number of Titles / Seasons")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

except KeyError as e:
    print("‚ùå Error:", e)
except Exception as e:
    print("‚ùå Unexpected error:", e)
````
